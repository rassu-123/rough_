#Clustering


##### Making Cluster based on 'Brand', 'Postal Code', 'Container Size'

```python
df_n1=df_new[['Material','Ship-to nu','Brand','Postal Code','Container Size']]
```

```python
df_cl1=df_new[['Brand','Postal Code','Container Size']]
df_cl1
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Use label encoder to make Brand value as numerical data

```python
df_cl1_try=df_cl1
```

```python
from sklearn.preprocessing import LabelEncoder
Brand = LabelEncoder()
input=df_cl1_try
input['Brand'] = Brand.fit_transform(input['Brand'])
```

```python
input
```

```python
input['Container Size'] = input['Container Size'].map(lambda x: x.lstrip('').rstrip('L'))
input
```

```python
# input['Container Size'].astype(float)
input['Container Size'] = input['Container Size'].str.replace(',', '').astype(float)
```

```python
input
```

##### X2 =input

```python
wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(input)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,6))    
plt.plot(range(1,11),wcss, linewidth=2, color="red", marker ="8")
plt.xlabel("K Value")
plt.xticks(np.arange(1,11,1))
plt.ylabel("WCSS")
plt.show()
```

##### Choose K value 3

```python
km2 = KMeans(n_clusters=3)
y2 = km2.fit_predict(input)
input["label"] = y2
df_n1["label"]=y2
#The data with labels
df_n1.head()
```

```python
# Calculate Silhoutte Score
from sklearn.metrics import silhouette_score
score = silhouette_score(input, km2.labels_, metric='euclidean')
#
# Print the score
#
print('Silhouetter Score: %.3f' % score)
```

```python
#3D Plot as we did the clustering on the basis of 3 input features
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(input.Brand[input.label == 0], input["Postal Code"][input.label == 0], input["Container Size"][input.label == 0], c='purple', s=60)
ax.scatter(input.Brand[input.label == 1], input["Postal Code"][input.label == 1], input["Container Size"][input.label == 1], c='red', s=60)
ax.scatter(input.Brand[input.label == 2], input["Postal Code"][input.label == 2], input["Container Size"][input.label == 2], c='blue', s=60)

ax.view_init(35, 185)
plt.xlabel("Brand")
plt.ylabel("Postal Code")
ax.set_zlabel('Container Size')
plt.show()
```

```python
from kmodes.kmodes import KModes
import matplotlib.pyplot as plt
```

```python
# Elbow curve to find optimal K
cost = []
K = range(1,11)
for num_clusters in list(K):
    kmode = KModes(n_clusters=num_clusters, init = "random", n_init = 5, verbose=1)
    kmode.fit_predict(input2)
    cost.append(kmode.cost_)
    
```

```python
plt.plot(K, cost, 'bx-')
plt.xlabel('No. of clusters')
plt.ylabel('Cost')
plt.title('Elbow Method For Optimal k')
```
